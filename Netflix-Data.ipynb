Country Column


```python
country = data['country'].value_counts().reset_index().sort_values(by='count',ascending = False).head(60)
country
```

This graph show the more shows is in US

Now determine the TVshow and movie in the every country

Top 10 Country


```python
country = data['country'].value_counts().reset_index().sort_values(by='count',ascending = False).head(10)
country
```

Number of Movies in Country


```python
plt.figure(figsize=(20,8))
plt.title('Top 10 countries')
sns.barplot(x='country',y='count',data=country)
plt.xlabel('Country')
plt.ylabel('No of Movies')
```

US have the greatest number of movie between top 10 Country


```python
top10_country=country.sort_values(by='count',ascending=False).head(10)
top10_country
```

```python
sns.barplot(data=top10_country,x='country',y='count',hue=data['type'],alpha=0.8,width=0.6)
plt.ylabel('Count')
plt.xlabel('Contries')
plt.title('Top 10 Countries')
plt.xticks(rotation=90)
plt.show()
```

Release Year


Distribution Show the more movie and TVshow is from 2010 to 2020


```python
data['release_year'].plot(kind='hist', bins=30)
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

Category


```python
category = data.groupby(['category', 'type'])['category'].value_counts().reset_index()
category
```

```python
top_10_listed_in_movie = category[category['type'] == 'Movie'].sort_values(by='count', ascending=False).iloc[0:10]
top_10_listed_in_movie
```

```python
top_10_listed_in_movie.plot(x='category', y='count', kind='bar')
plt.title('Top 10 Listed Genres for Movies')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
